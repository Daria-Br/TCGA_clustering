# LUAD KRAS EGFR ALK Mutational Status



```python
import numpy as np
import pandas as pd
```

```python
df_mut = pd.read_csv("KRAS_EGFR_ALK_mut_stat_LUAD.tsv", sep= "\t", index_col= 0)
df_clusters = pd.read_excel("LUAD_Table_for_decon.xlsx", sheet_name = 3, index_col = 0)
```

```python
df_mut
```

```python
df_clusters
```

```python
df = df_clusters.join(df_mut)
df
```

```python
KRAS = df.groupby("Cluster")["KRAS"].value_counts()
KRAS
```

```python
EGFR = df.groupby("Cluster")["EGFR"].value_counts()
EGFR
```

```python
ALK = df.groupby("Cluster")["ALK_translocation"].value_counts()
ALK
```

```python
with pd.ExcelWriter('LUAD_mut_analysis.xlsx') as writer:
    KRAS.to_excel(writer, sheet_name='KRAS')
    EGFR.to_excel(writer, sheet_name='EGFR')
    ALK.to_excel(writer, sheet_name="ALK_translocation")
```
